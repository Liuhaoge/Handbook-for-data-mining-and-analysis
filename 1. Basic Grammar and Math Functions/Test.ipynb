```python
from bs4 import BeautifulSoup
import requests
html = requests.get('http://www.polyv.net')
 
#使用content属性获取页面的源页面
#使用BeautifulSoap解析，把内容传递到BeautifulSoap类
soup = BeautifulSoup(html.content,'lxml')

soup
```

```python
links = soup.find_all('p')

links
```

```python
for i in links:
    print(i.get_text())
```
