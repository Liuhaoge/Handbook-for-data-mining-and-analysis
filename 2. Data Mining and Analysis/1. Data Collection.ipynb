```python
# Application：MonteCarlo Method calculating π
from numpy.random import random
from math import sqrt
DARTS = 1000000
hits = 0.0

for i in range(1, DARTS+1):
    x, y = random(), random()
    dist = sqrt(x ** 2 + y ** 2)
    if dist <= 1.0:
        hits = hits + 1
pi = 4 * (hits/DARTS)
print("The value of π is %f" %pi)
```

### random number by distrubtion

```python
import numpy as np

mu, sigma, num = 1,3,10  # expectation: 1, deviation: 3, number: 10

np.random.normal(mu, sigma, num) # normal distribuation
```

```python
scale = 50
size = 20

np.random.exponential(scale, size)

lam=15
np.random.poisson(lam, size) 

low,high = 10,50
np.random.uniform(low, high, size)
```

# Sampling 

```python
df = pd.DataFrame(np.arange(5 * 4).reshape((5, 4)))
df

# Method one
sampler = np.random.permutation(3)
df.take(sampler)
```

```python
# Method two
df.take(np.random.permutation(len(df))[:3])
```

```python
# Method three
bag = np.array([5, 7, -1, 6, 4])
sampler = np.random.randint(0, len(bag), size=10)
draws = bag.take(sampler)
draws
```

### CSS selector： select方法

```python
html='''
<div class="panel">
    <div class="panel-heading">
        <h4>Hello</h4>
    </div>
    <div class="panel-body">
        <ul class="list" id="list-1">
            <li class="element">Foo</li>
            <li class="element">Bar</li>
            <li class="element">Jay</li>
        </ul>
        <ul class="list list-small" id="list-2">
            <li class="element">Foo</li>
            <li class="element">Bar</li>
        </ul>
    </div>
</div>
'''
soup = BeautifulSoup(html, 'lxml')
```

```python
print(soup.select('.panel .panel-heading'))  #如果是class，则在前面加点.
```

```python
print(soup.select('ul li'))     #如果是标签
```

```python
print(soup.select('#list-2 .element'))      #如果是id，则在前面加#

print(soup.select('#list-2 .element')[0])
```

```python
#获取内容
for li in soup.select('li'):
    print(li.get_text())
```
