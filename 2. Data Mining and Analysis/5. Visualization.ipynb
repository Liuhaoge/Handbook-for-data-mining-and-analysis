### Figures and Subplots
`figure` 对象是最外层的绘图单位，默认是以 `1` 开始编号（**MATLAB** 风格，`Figure 1, Figure 2, ...`），可以用 `plt.figure()` 产生一幅图像，除了默认参数外，可以指定的参数有：

- `num` - 编号
- `figsize` - 图像大小
- `dpi` - 分辨率
- `facecolor` - 背景色
- `edgecolor` - 边界颜色
- `frameon` - 边框

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1,t2 = np.arange(0.0, 5.0, 0.1), np.arange(0.0, 5.0, 0.02)

plt.figure(1)
plt.subplot(311)    # Three rows, one column, the first chart
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')

plt.subplot(312)   # Three rows, one column, the second chart
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')

plt.subplot(313)   # Three rows, one column, the second chart
plt.plot(t2, np.exp(t2), 'cp')
plt.show()
```

### Plot with words

在 `matplotlib.pyplot` 中，基础的文本函数如下：

- `text()` 在 `Axes` 对象的任意位置添加文本
- `xlabel()` 添加 x 轴标题
- `ylabel()` 添加 y 轴标题
- `title()` 给 `Axes` 对象添加标题
- `figtext()` 在 `Figure` 对象的任意位置添加文本
- `suptitle()` 给 `Figure` 对象添加标题
- `anotate()` 给 `Axes` 对象添加注释（可选择是否添加箭头标记）

```python
# -*- coding: utf-8 -*-
# plt.figure() 返回一个 Figure() 对象, figure可以设定画布的大小
fig = plt.figure(figsize=(8, 6))

# 设置这个 Figure 对象的标题
# 事实上，如果我们直接调用 plt.suptitle() 函数，它会自动找到当前的 Figure 对象
fig.suptitle('bold figure suptitle', fontsize=14, fontweight='bold')

# Axes 对象表示 Figure 对象中的子图
# 这里只有一幅图像，所以使用 add_subplot(111)
ax = fig.add_subplot(111)
fig.subplots_adjust(top=0.85)

# 可以直接使用 set_xxx 的方法来设置标题
ax.set_title('axes title')
# 也可以直接调用 title()，因为会自动定位到当前的 Axes 对象
# plt.title('axes title')

ax.set_xlabel('xlabel')
ax.set_ylabel('ylabel')

# 添加文本，斜体加文本框
# 前两个参数指定text的位置
ax.text(3, 8, 'boxed italics text in data coords', style='italic',
        bbox={'facecolor':'red', 'alpha':0.5, 'pad':10})

# 数学公式，用 $$ 输入 Tex 公式
ax.text(2, 6, r'an equation: $E=mc^2$', fontsize=15)

# Unicode 支持
# ax.text(3, 2, unicode('unicode: Institut f\374r Festk\366rperphysik', 'latin-1'))

# 颜色，对齐方式
ax.text(0.95, 0.01, 'colored text in axes coords',
        verticalalignment='bottom', horizontalalignment='right',
        transform=ax.transAxes,
        color='green', fontsize=15)

# 注释文本和箭头
ax.plot([2], [1], 'o')
ax.annotate('annotate', xy=(2, 1), xytext=(3, 4),
            arrowprops=dict(facecolor='black', shrink=0.05))

# 设置显示范围
ax.axis([0, 10, 0, 10])

plt.show()
```

### Hist


```python
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000) 

# the histogram of the data: x is the number, y is the count or probability of x
n, bins, patches = plt.hist(x, bins=50, density=1, histtype='bar', facecolor='g', edgecolor='#000000', alpha=0.75) # density=1 means y is probability

# add a 'best fit' line
from scipy.stats import norm

y=norm.pdf(bins,mu,sigma)
plt.plot(bins, y, 'r--')

plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60, .025, r'$\mu=100,\ \sigma=15$')
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
```

```python
import scipy.stats

fig, axs = plt.subplots(figsize = (5, 5))
N_numbers = 1000
N_bins = 100
k = 10000

mu1, mu2 = 0, 0
sigma1, sigma2 = 2, 4
mu = np.array([mu1, mu2])
Sigma = np.array([[sigma1, -1.5], [-1.5, sigma2]])

mn = sp.stats.multivariate_normal(mean=mu, cov=Sigma)
X = mn.rvs(size=k)           
axs.hist2d(X[:,0], X[:,1], bins=N_bins, density=False, cmap='plasma')
```

### Bar

```python
means_men, means_women = [20, 35, 30, 35, 27],[25, 32, 34, 20, 25]  
std_men, std_women = [2, 3, 4, 1, 2],[3, 5, 2, 3, 3]  # 方差

n_groups = 5
index = np.arange(n_groups)
bar_width = 0.35

opacity = 0.4  # 透明度

rects1 = plt.bar(index, means_men, bar_width,alpha=opacity,color='b',yerr=std_men,label='Men')  # yerr是方差
rects2 = plt.bar(index + bar_width, means_women, bar_width,alpha=opacity,color='g',yerr=std_women,label='Women')

plt.xlabel('Group')
plt.ylabel('Scores')
plt.title('Scores by group and gender')
plt.xticks(index + bar_width, ('A', 'B', 'C', 'D', 'E'))
plt.legend()
```

```python
# 横向柱状图 barh
means_men, means_women = [20, 35, 30, 35, 27],[25, 32, 34, 20, 25]  
std_men, std_women = [2, 3, 4, 1, 2],[3, 5, 2, 3, 3]  # 方差

n_groups = 5
index = np.arange(n_groups)
bar_width = 0.35

opacity = 0.9  # 透明度

rects1 = plt.barh(index, means_men, bar_width,alpha=opacity,color='b',label='Men')  # yerr是方差
rects2 = plt.barh(index + bar_width, means_women, bar_width,alpha=opacity,color='g',label='Women')

plt.ylabel('Group')
plt.xlabel('Scores')
plt.title('Scores by group and gender')
plt.yticks(index + bar_width, ('A', 'B', 'C', 'D', 'E'))
plt.legend()
```

```python
from pylab import *

n = 12
X = np.arange(n)
Y1 = (1-X/float(n)) * np.random.uniform(0.5,1.0,n)
Y2 = (1-X/float(n)) * np.random.uniform(0.5,1.0,n)

bar(X, +Y1, facecolor='#9999ff', edgecolor='white', label="Example one")
bar(X, -Y2, facecolor='#ff9999', edgecolor='white', label="Example two")

for x,y in zip(X,Y1):
    text(x+0.4, y+0.05, '%.2f' % y, ha='center', va= 'bottom')

plt.legend()
ylim(-1.25,+1.25)
show()
```

### Contour

```python
def f(x,y): return (1-x/2+x**5+y**3)*np.exp(-x**2-y**2)

n = 256
x = np.linspace(-3,3,n)
y = np.linspace(-3,3,n)
X,Y = np.meshgrid(x,y)

contourf(X, Y, f(X,Y), 8, alpha=.75, cmap='jet')
C = contour(X, Y, f(X,Y), 8, colors='black', linewidth=2.0)
show()
```

### 3D plot

```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D

fig = figure()
ax = Axes3D(fig)
X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='hot')

show()
```
